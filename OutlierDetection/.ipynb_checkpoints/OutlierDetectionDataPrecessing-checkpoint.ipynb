###  读入所有数据并保存到字典中
>从数据中可以发现，很多全是NaN值，这就需要进行数据的清洗， 处理掉这些缺失值
### 标签标注
> 现在的数据并没有标签， 只是有了温度和盐度的测量结果和是不是异常值，后面的temp_res和sail_res就是标注的温度和盐度的异常与否，如果想获得标签， 需要两步： 
>> * 第一步，需要先处理掉sail_res里面的NaN值，之所以这里会出现NaN，是因为有些数据盐度没有存在， 这时候处理的时候，由于没有测盐度，所以默认这样的样本是没有异常的， 统一先把NaN值用1进行填充
>> * 设置sail_res里面的数据格式，这里面通过上面的dtypes属性的查看，目前是float型的，需要强转成int
>> * 然后就可以结合者这两列，进行与运算，获得最后的标签label，单独设置一列在后面
>
> PS：这里做标签标注的时候，由于拿到的是师哥标注好的样本，所以在这里先直接用，但是师哥标注的时候，简单化了一些假设，把可能正确的数据归为正确，可能错误的数据归为错误， 然后只有1和0<br>
我的想法是目前先用这个跑跑看，后期的话如果再改进，这里可以是一个改进点， 就是可能正确和错误的那些不能这样划分，如果少的话可以当做未知数据进行模型预测出来

```python
data = {}
datapath = 'data/'
datalist = os.listdir(datapath)
print(datalist)


for filename in datalist:
    tempdata = pd.read_csv('data/' + filename)
    tempdata['sali_res'] = tempdata['sali_res'].fillna(1.0)
    tempdata['temp_res'] = tempdata['temp_res'].fillna(1)
    tempdata['sali_res'].astype(int)
    tempdata['label'] = tempdata['temp_res'] * tempdata['sali_res']
    tempdata['label'].astype(int)      # 标签转换成整型
    
    data[filename[0:4]] = tempdata
    print(filename)
```

> 通过上面的代码，已经把data下面的所有数据存放到了一个字典里面，由于每个文件的特征长度都是不一样的，所以没办法直接合并，只能先大体分析一下数据，比如负样本和正样本的个数，然后初步思路是特征提取的时候，提取和异常样本有关的特征，因为我觉得，是因为这些特征才导致了探测器产生了异常，能分析的特征感觉也就是这几个。
### 数据分析

```python
"""先统计一下，样本里面正负样本的个数"""
TrueSampleNum = 0
FalseSampleNum = 0
for v in data.values():
    Labelcount = v['label'].value_counts()
    print(Labelcount)
    index = Labelcount.index
    TrueSampleNum += Labelcount[1.0]
    if 0.0 in index:
        FalseSampleNum += Labelcount[0.0]

print("样本中非异常样本的个数：", TrueSampleNum)
print("样本中异常样本的个数： ", FalseSampleNum)
print(TrueSampleNum/FalseSampleNum)      # 正负样本80：1的比例
```

### 数据处理 - 提取数据
> 接下来需要提取特征， 这里的一个难题就是这些数据的深度和盐度都是变长的， 并且大部分都是空值， 并且列还不一样，所以我想
>> * 第一步， 就是去掉每个数据集里面的整列全是空的列
>> * 第二步， 把所有异常值筛选出来，那些异常值里面不是空的那些特征值的参考
>> * 第三步， 把第二部筛选出来的特征进行合并，然后基于这个统一所有的特征长度
>> * 第四步， 处理缺失值， 再进行考虑后面的。

#### 第一步， 去掉每个数据集里面的整列全是NaN的列

```python
# print(data.keys())
for key in data.keys():
    data1 = data[key]
    print(data1.shape)
    nullcount = data1.isnull().sum()
    # 删除全是空行的列
    for i in data1.columns:
        if nullcount[i] == data1.shape[0]:
            del data1[i]
    print(data1.shape)
```

#### 第二步， 提取有用的特征
> 我觉得， 既然是异常值检测， 那么是什么导致会出现异常值呢？  我分析了以下特征：
>> * 1. 时间是一个特征， 因为早上和下午时间的不同，可能会影响水温，进而可能会导致出现异常，但是事件这个，可能不太好取数值，后面再说
>> * 2. 经度和纬度是特征，因为不同的经纬度代表着地点不同， 可能会产生异常值
>> * 3. 仪器代码是一个特征，不同的仪器可能会产生异常
>> * 4. 水的深度是一个特征， 不同水深可能发生异常
>> * 5. 而反应异常不异常的是温度和盐度， 所以这两个是检测指标， 也是特征
> 
> 但是就像后面的4，5可能会有多组， 给定的输入肯定是时间，地点，仪器号，然后水深多少， 温度值时多少， 盐度值多少， 所以前面三个可以正确的看成特征1，2， 3，但是4，5这个需要再进一步处理提取。
>
><br>
>
>由于去掉全是空值的特征列之后还有列数很多的，我想进一步进行提取，我觉得，既然是异常值检测，那么就需要先基于异常值， 所提取的水深应该是基于异常值包含的里面提取的，否则感觉学习并没有意义，所以统一特征部分我想这样做：
>> * 1. 先建一个features空列表存异常样本有的特征
>> * 2. 遍历所有数据集，对于每一个数据集，得先提取出它的异常样本数据集，遍历这个数据集的每一列，去掉全是空的，这种也没有意义，然后把这样的特征存到features里面
>> * 3. 这样统一的特征就是features里面的特征，接下来进行筛选， 再遍历每一个数据集，然后遍历每个数据集的每一列，如果在features里面，就保存，否则删掉
>> * 4. 构造新的统一化数据集， 创建一个data_new的DataFrame对象，遍历所有数据集，先把前1.2.3个特征列加入， 然后遍历features，从数据集中对应的列加入新的数据集，一轮之后， 进行行堆集。 这样的话数据集就统一了顺序和特征。
>
> 有了统一化的数据集之后，还是先处理缺失值，分析数据，这时候就可以画图像来，如果4，5特征列还是很多的话，我想再采用师哥说的方式进行压缩， 因为我觉得对于变长数据，rnn的mask是可以处理，但是我查资料发现mask好像不是处理这个东西的，并且如果列数不同一，空值太多的话，我不会直接用rnn的mask进行处理，所以我想用我的方式先看看效果。

```python
# 筛选出所有的异常值以及异常值所在的特征
features = []
for key in data.keys():
    tempdata = data[key]
    outlier = tempdata[tempdata['label']==0.0]   # 提取异常样本
    
    # 遍历异常样本的每一列， 去掉所有行都是空的， 这些就是异常样本提供的参考的有用的特征,加入features
    for col in outlier.columns:
        if outlier[col].isnull().sum() == outlier.shape[0]:
            del outlier[col]
            
    for col in outlier.columns:
        if col not in features:
            features.append(col)

print(len(features))   
```

```python
print(features)
```

> 结果features的列数还是太多了， 没法处理， 所以我下面的思路就是再进行列的缩减，所以需要再从原来的基础上找到异常样本的那个引起异常的那个特征
>> * 遍历所有的原始数据集，就是没转成CSV的那个， 从那里面一行行的读取， 筛选出仪器可能出错的那些深度， 那些作为关键的深度特征，（即使后面真的无监督处理，这些也是重点的计算均值的那些列。）
>> * 提取出那些深度之后，遍历features列，看看是否在关键的深度特征里面，如果没有，删去一些特征（注意这个地方最好备份一份）

```python
origdatapath = 'origdata/'
origdatalist = os.listdir(origdatapath)
print(origdatalist)
key_sample = []
for origdataname in origdatalist:
    with open('origdata/' + origdataname, 'r') as f:
        line = f.readline()
        while line:
            line = f.readline()
            line_split = line.strip().split(' ')
            if line_split[0] == '2':
                if (line_split[3]=='2' or line_split[3]=='3' or line_split[3]=='4'):
                    key_sample.append(line_split)
                if (line_split[5]=='2' or line_split[5]=='3' or line_split[5]=='4'):
                    key_sample.append(line_split)
print(len(key_sample))
```

> key_sample里面就是那些出现异常或者可能出现异常的那些样本中的某一个深度的值， 含义：<br>数据标识，深度，温度值，温度质控符，盐度值，盐度质控符
>
> 上面的9432就是说，有这么多个数据发生了错误或者可能存在错误，这些是我们想研究的， 接下来，需要遍历查看一下这个二维数组，筛选出这些关键的深度。但是深度里面可能是温度发生错误，可能是盐度发生错误，这里又需要做一个区分

```python
# 简单的查看一下前50个
for i in range(len(key_sample[:50])):
    print(key_sample[i])
```

##### 进一步筛选
>尝试筛选出关键的深度，并且根据具体是温度还是盐度，组合成和features中列一样的特征列，才能进一步缩减features列的数量
>
><br>
>
> 下面理理上面的思路： 特征提取部分，是首先把异常的样本筛选出来，把异常样本包含的所有特征去掉全是空值的列，其他的放到了features列表中， 但是我发现这样提取之后，features里面的列还是太多，9700多个， 对于后面的筛选等造成困难，所以才有了进一步筛选，这次筛选是选出那些使得异常样本成为异常的那些特征，也就是从异常样本所包含的所有深度列中，选出使得仪器出错的那些深度列，进行对features的列的缩减。
>
><br>
>
> 处理key_sample这个二维列表
>> * 遍历列表，获取每一行，判断是[3]等于2，3，4还是[5]等于2，3，4
>>> * 如果是[3]等于2，3，4， 那么构造的特征就是"temperature"+[1]
>>> * 如果是[5]等于2，3，4， 那么构造的特征就是"salinity"+[1]（[1]表示深度）
>> * 把上面构造的特征再存放到一个空的列表中，然后和features做比对



```python
features_mask = []
# 遍历key_samples，获取到每一行
for v in key_sample:
    if v[3]=='2' or v[3]=='3' or v[3]=='4':
        new_feature = "temperature" + v[1]
        if new_feature not in features_mask:
            features_mask.append(new_feature)
    if v[5]=='2' or v[5]=='3' or v[5]=='4':
        new_feature = "salinity" + v[1]
        if new_feature not in features_mask:
            features_mask.append(new_feature)
```

```python
# print(features_mask)
print(len(features_mask))
# print(features_mask[:50])
# print(features[:50])

"""接下来进行筛选，得到处理后的新特征"""
features_new = ['time', 'latitude', 'longitude', 'instrument']
for i in features:
    if i in features_mask:
        features_new.append(i)

features_new.append('label')
print(len(features_new))
print(features_new[:50])

"""这里记录一下：  我发现了一个问题， 这两个数字相差太大，这说明了一些问题， 我是直接使用师哥标注的数据， 根据标注的数据，我先把异常的
也就是0的那些样本的特征全部合在了一块， 构成features， 然后由于features太大，我想进一步筛选，就遍历了一遍数据集，把可能出现问题的
那些深度特征筛选出来， 这个就是features_mask,理论上讲，如果之前features处理正确的话，features中的特征是包含features_mask里面的
特征的，但是，我又遍历了一遍features， 发现里面的特征数目从3146减到来2928， 这说明features里面并不是包含features_mask的，所以这里
可能我筛选出了问题，或者师哥标注出了问题， 这个先用2928算着，看看效果吧"""
```

#### 第三步 构造统一化数据集
> * 首先遍历所有的数据集， 然后遍历features_new，根据这个顺序把数据集加入到新的data_new中，
> * 遍历完一个之后，采用行堆集的方式，把所有的都放上来

```python
print(data['1965'].shape)
data_new = np.zeros([1, 2929])
print(data_new.shape)
for v in data.values():
    tempdata = pd.DataFrame()
    for feature in features_new:
        if feature in v.columns:
            tempdata = pd.concat([tempdata, v[feature]], axis=1)
        else:
            tempdata[feature] = np.nan
    print(tempdata.values.shape)
    data_new = np.vstack((data_new, tempdata.values))
        
print(data_new.shape)

# 备份一份数据
data_new_data = data_new
```
