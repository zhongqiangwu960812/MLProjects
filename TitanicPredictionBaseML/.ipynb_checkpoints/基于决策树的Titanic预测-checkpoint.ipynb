## Titanic号决策树实例
下面开始使用决策树完成Titanic号的存活预测问题
> 数据预处理前期工作：
>> * 筛选特征值，丢掉不需要的特征数据
>> * 对性别进行二值化处理（转换为0和1）
>> * 港口转换成数值型数据
>> * 处理缺失值（如年龄，有很多缺失值）

## 读取数据集

```python
"""导入用到的包"""
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder

from sklearn.model_selection import train_test_split#数据拆分的函数
```

```python
"""导入数据集"""
def read_dataset(fname):
    # 指定第一列作为行索引
    data = pd.read_csv(fname, index_col=0)
    
    # 丢弃无用的数据
    data.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
    
    # 处理性别数据
    lab = LabelEncoder()
    data['Sex'] = lab.fit_transform(data['Sex'])
#     lables = data['Sex'].unique().tolist()          # 变成列表
#     data['Sex'] = [*map(lambda x: lables.index(x), data['Sex'])]
    
    
    # 处理登船港口数据
    #data['Embarked'] = lab.fit_transform(data['Embarked'])
    labels = data['Embarked'].unique().tolist()
    data['Embarked'] = data['Embarked'].apply(lambda n: labels.index(n))
    print(data['Embarked'].dtype)
    
    # 处理缺失数据，填充为0
    data = data.fillna(0)
    return data

train = read_dataset("dataSet/train.csv")
```

## 拆分数据集
>把Survived列提取出来作为标签，然后在元数据集中将其丢弃。同时拆分数据集和交叉验证数据集

```python
y = train['Survived'].values    # 标签
x = train.drop(['Survived'], axis=1).values    # 特征

X_train, X_valid, y_train, y_valid = train_test_split(x, y, test_size=0.2, random_state=10)  #随机选择百分之二十作为验证测试用
print("X_train_shape:", X_train.shape, " y_train_shape:", y_train.shape)
print("X_test_shape:", X_valid.shape,"  y_test_shape:", y_valid.shape)
```

## 拟合数据集--建模拟合

```python
clf = DecisionTreeClassifier()#实例化
clf.fit(X_train, y_train)#训练
print("train score:", clf.score(X_train, y_train))#过拟合
print("test score:", clf.score(X_valid, y_valid))#
```

## 优化参数模型

通过max_depth参数来优化模型

从以上输出数据可以看出，针对训练样本评分很高，但针对测试数据集评分较低。很明显这是过拟合的特征。解决决策树过拟合的方法是剪枝，包括前剪枝和后剪枝。但是sklearn不支持后剪枝，这里通过max_depth参数限定决策树深度，在一定程度上避免过拟合。

这里先创建一个函数使用不同的模型深度训练模型，并计算评分数据。

```python
def cv_score(d):
    clf = DecisionTreeClassifier(max_depth=d)
    clf.fit(X_train, y_train)
    return (clf.score(X_train, y_train), clf.score(X_valid, y_valid))
```

```python
"""尝试自己去搜索参数"""
depths = np.arange(1, 10)   # 取1到9
scores = [cv_score(d) for d in depths]
tr_scores = [s[0] for s in scores]
te_scores = [s[1] for s in scores]

# 找出交叉验证数据集评分最高的索引
tr_best_index = np.argmax(tr_scores)
te_best_index = np.argmax(te_scores)

print("bestdepth: ", te_best_index+1, "bestdepth_score: ", te_scores[te_best_index])

"""这里由于以上train_test_split方法对数据切分是随机打散的，造成每次用不同的数据集训练模型总得到不同的最佳深度。
这里写个循环反复测试，最终验证这里看到最佳的分支深度为3出现的频率最高，初步确定6为深度模型最佳。"""
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
"""下面尝试可视化结果，通过图的形式一目了然"""

depths = np.arange(1, 10)
plt.figure(figsize=(10, 4), dpi=111)       # 图的大小和缩放比例
plt.grid()   # 显示网格
plt.xlabel("max depth of decision tree")
plt.ylabel("Scores")
plt.plot(depths, te_scores, label="test_scores")
plt.plot(depths, tr_scores, label="train_scores")
plt.legend()   # 显示图例
```

## 通过min_impurity_decrease来优化模型--剪枝的意思

这个参数用来指定信息墒或者基尼不纯度的阀值，当决策树分裂后，其信息增益低于这个阀值时则不再分裂。

```python
def minsplit_score(val):
    clf = DecisionTreeClassifier(criterion='gini', min_impurity_decrease=val)  # 模型的参数
    clf.fit(X_train, y_train)
    return (clf.score(X_train, y_train), clf.score(X_valid, y_valid))  

# 指定参数范围， 分别训练模型并计算得分
vals = np.linspace(0, 0.2, 100)   # 平均分成一百份
scores = [minsplit_score(v) for v in vals]
tr_scores = [s[0] for s in scores]
te_scores = [s[1] for s in scores]

bestmin_index = np.argmax(te_scores)    ## 选出最大的那个评分的索引
bestscore = te_scores[bestmin_index]
print("bestmin: ", vals[bestmin_index])
print("bestscore:", bestscore)
```

```python
"""下面尝试可视化"""
plt.figure(figsize=(10, 4), dpi=120)
plt.grid()
plt.xlabel("min_impurity_decrease")
plt.ylabel("Scores")
plt.plot(vals, te_scores, label="test_scores")
plt.plot(vals, tr_scores, label="train_scores")
plt.legend()
```

值此为我们找到了两个参数,最佳深度depth=6 和最佳min_impurity_decrease=0.002，下面我来用两个参数简单对模型进行测试：

```python
from sklearn import metrics
```

```python
model = DecisionTreeClassifier(max_depth=6, min_impurity_decrease=0.002)
model.fit(X_train, y_train)

print("tees_score: ", model.score(X_valid, y_valid))

y_pred = model.predict(X_valid)

print("查准率：", metrics.precision_score(y_valid, y_pred))
print("召回率: ", metrics.recall_score(y_valid, y_pred))
print("F1_score: ", metrics.f1_score(y_valid, y_pred))
```

准确率（accuracy）计算公式如下所示：

准确率表示预测正确的样本（TP和TN）在所有样本（all data）中占的比例。

　　在数据集不平衡时，准确率将不能很好地表示模型的性能。可能会存在准确率很高，而少数类样本全分错的情况，此时应选择其它模型评价指标。
  



发现一个问题：每次运行的结果都不一样，这和我们所需要得一个稳定的结果有很大的差别呢，怎么办？

1、数据不稳定：–> 每次重新分配训练集测试集，原参数就不是最优了。 解决办法是多次计算求平均值。

2、不能一次选择多个参数：–> 想考察max_depth和min_impurity_decrease两者结合起来的最优参数就没法实现。

解决方案：所幸scikit-learn在sklearn.model_selection包提供了大量的模型选择和评估的工具供我们使用。针对该问题可以使用GridSearchCV类来解决。

## 利用GridSearchCV求最优参数--实现找到多个最优参数

```python
from sklearn.model_selection import GridSearchCV   # 网格搜索
```

```python
thresholds = np.linspace(0, 0.2, 50)
param_grid = {'min_impurity_decrease': thresholds, "max_depth":np.arange(1,10)}

clf = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5)   # 算法， 参数， 交叉验证的次数
clf.fit(x, y)

print("best_params:{0} \n best_score:{1}".format(clf.best_params_, clf.best_score_))
```

参数解读：

1、关键字参数param_grid是一个字典，字典的关键字对应的值是一个列表。GridSearchCV会枚举列表里所有值来构建模型多次计算训练模型，并计算模型评分，最终得出指定参数值的平均评分及标准差。

2、关键参数cv，用来指定交叉验证数据集的生成规则。这里cv=5表示每次计算都把数据集分成5份，拿其中一份作为交叉验证数据集，其他作为训练集。最终得出最优参数及最优评分保存在clf.best_params_和clf.best_score_里。

3、此外clf.cv_results_里保存了计算过程的所有中间结果。

```python
clf.cv_results_
```

```python
def plot_curve(train_sizes, cv_results, xlabel):
    train_scores_mean = cv_results['mean_train_score']
    train_scores_std = cv_results['std_train_score']
    test_scores_mean = cv_results['mean_test_score']
    test_scores_std = cv_results['std_test_score']
    
    plt.figure(figsize=(6, 4), dpi=120)
    plt.title('parameters turning')
    plt.grid()
    plt.xlabel(xlabel)
    plt.ylabel('score')
    plt.fill_between(train_sizes,  # 范围填充
                    train_scores_mean - train_scores_std,
                    train_scores_mean + train_scores_std,
                    alpha=1, color="r")
    plt.fill_between(train_sizes,
                    test_scores_mean - test_scores_std,
                    test_scores_mean + test_scores_std,
                    alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, '.--', color="r", label="Training score")
    plt.plot(train_sizes, test_scores_mean, '.-', color='g', label='Cross-validation score')
    
    plt.legend(loc='best')
```

```python
thresholds = np.linspace(0, 0.2, 50)
# Set the parameters by cross-validation
param_grid = {'min_impurity_decrease': thresholds}

clf = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5)
clf.fit(x, y)
print("best param: {0}\nbest score: {1}".format(clf.best_params_, 
                                                clf.best_score_))

print(clf.cv_results_)
plot_curve(thresholds, clf.cv_results_, xlabel='gini thresholds')
```
