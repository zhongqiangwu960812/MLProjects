## 先来瞄一眼数据集

```python
"""导入包"""
from sklearn.datasets import fetch_mldata
```

```python
mnist = fetch_mldata('MNIST original', data_home='./data/')
mnist
```

```python
X, y = mnist["data"], mnist["target"]  #特征以及标签
X.shape, y.shape
```

```python
#先来看一下数据集，使用matplotlib进行可视化，这里要理解把向量转化成矩阵才能显示
import matplotlib
import matplotlib.pyplot as plt#可视化包
 
digitshow = X[40]#第50000个实例,样本
digitshow= digitshow.reshape(28, 28)#同reshape函数转化
plt.imshow(digitshow, cmap = matplotlib.cm.binary,interpolation="nearest")#传入图片，颜色，位置
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier#导入多层感知器分类机BP网络
X = X / 255.#数据归一化

#  划分数据集也可以用我们之前用的train_test_split函数
##切片的形式
X_train, X_test = X[:60000], X[60000:]#选取前六万张作为训练集
y_train, y_test = y[:60000], y[60000:]#后一万张作为测试集

mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=400, alpha=1e-4,
                    solver='adam', verbose=10, tol=1e-4, random_state=1)
# mlp = MLPClassifier(hidden_layer_sizes=(100,), activation="relu",max_iter=400, alpha=1e-4,
#                     solver='adam', verbose=10, tol=1e-4, random_state=1)
"""
参数说明：hidden_layer_sizes=(50,)：第一层784个神经元，隐藏层50个神经元，输出层10个神经元（十个类别）
activation：代表的是每个神经元的激活函数：logistic是sigmoid激活函数，此外还有relu，tanh等常用的激活函数
max_iter=10:迭代次数
solver='sgd'：优化器
learning_rate_init=0.1：初始学习率
alpha=1e-4：步长
tol=1e-4：改善指标，也算是一个终止条件
verbose=10：打印日志训练的日志
"""
##详细的参数说明请看此链接
#https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html#sklearn.neural_network.MLPClassifier
mlp.fit(X_train, y_train)#模型的训练
print("Training set score: %f" % mlp.score(X_train, y_train))#对训练集评分
print("Test set score: %f" % mlp.score(X_test, y_test))#训练好的模型对测试集评分




# fig, axes = plt.subplots(4, 4)
# # use global min / max to ensure all weights are shown on the same scale
# vmin, vmax = mlp.coefs_[0].min(), mlp.coefs_[0].max()
# for coef, ax in zip(mlp.coefs_[0].T, axes.ravel()):
#     ax.matshow(coef.reshape(28, 28), cmap=plt.cm.gray, vmin=.5 * vmin,
#                vmax=.5 * vmax)
#     ax.set_xticks(())
#     ax.set_yticks(())

# plt.show()
```
